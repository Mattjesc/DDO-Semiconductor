DATA PREPARATION STAGE

```python
# Step 3: Handling Missing Values

import pandas as pd

# Load the original dataset (before any preparation)
# Justification: The dataset is loaded again to ensure that the raw, unaltered data is used for handling missing values.
# This is crucial because previous steps (like initial exploration) might have altered the dataset, and we want to start 
# with a clean slate for handling missing data.
input_path = '../data/FeatureSelection.csv'
df = pd.read_csv(input_path)

# Handling Missing Values
# Justification: Checking for missing values is an essential step in data preparation. Missing data can lead to biased estimates,
# reduced statistical power, or incorrect conclusions if not handled properly. This step identifies which columns have missing data.
missing_values = df.isnull().sum()  # Sum up missing values in each column.
print("Missing Values in Each Column:\n", missing_values[missing_values > 0])  # Print columns that have missing values.

# Impute missing values with the median for numeric columns
# Justification: Imputation is used to fill in missing data. The median is chosen for numeric columns because it is robust to outliers 
# (unlike the mean) and provides a good central tendency estimate for skewed distributions. This ensures that the missing values 
# are filled in a way that doesn't overly influence the data distribution.
numeric_columns = df.select_dtypes(include='number').columns  # Select numeric columns for imputation.
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())  # Fill missing values with the median.

# Verify that there are no more missing values
# Justification: After imputation, it's crucial to verify that all missing values have been appropriately handled. 
# This check ensures that the dataset is now complete and ready for further processing.
print("\nMissing Values After Imputation (Numeric Columns Only):\n", df[numeric_columns].isnull().sum().sum())  # Should print 0 if all missing values are filled.

# Save the dataset after handling missing values
# Justification: The cleaned dataset is saved to a new file to preserve the changes made. This also ensures that subsequent steps 
# in the data pipeline use the correct version of the data, which now has no missing values.
output_path = '../Outputs_FeatureSelection/FeatureSelection_Prep1.csv'  # Define the path for saving the cleaned dataset.
df.to_csv(output_path, index=False)  # Save the dataset without the index for consistency in future steps.

print(f"Dataset after handling missing values saved to {output_path}")  # Inform the user that the dataset has been successfully saved.
```

```python
# Step 4: Normalization and Standardization

from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the dataset from the previous step (Prep1)
# Justification: The dataset prepared in the previous step, where missing values were handled, is loaded for normalization
# and standardization. These steps are critical to ensure that the data is on the same scale, which is important for many 
# machine learning models that assume input features are on a similar scale.
input_path = '../Outputs_FeatureSelection/FeatureSelection_Prep1.csv'
df = pd.read_csv(input_path)

# Normalize and standardize the numeric features
# Justification: Standardization transforms the data to have a mean of 0 and a standard deviation of 1. This is crucial 
# for algorithms like Logistic Regression, SVMs, and Neural Networks, which are sensitive to the scales of input data.
# It ensures that each feature contributes equally to the model, preventing features with larger scales from dominating the learning process.
numeric_columns = df.select_dtypes(include='number').columns  # Select numeric columns for normalization and standardization.
scaler = StandardScaler()  # Initialize the scaler which standardizes the features by removing the mean and scaling to unit variance.
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])  # Apply the standardization to all numeric columns.

# Revert the normalization of 'Pass/Fail' if it was mistakenly normalized
# Justification: The 'Pass/Fail' column is the target variable and should not be normalized or standardized as it's categorical.
# Reverting any accidental transformation ensures the integrity of the target variable, which is essential for model training.
df['Pass/Fail'] = pd.read_csv(input_path)['Pass/Fail']  # Re-load the original 'Pass/Fail' column to ensure it remains unaltered.

# Display the first few rows of the processed data to verify normalization
# Justification: Displaying the first few rows after normalization provides a quick check to ensure the transformation was applied correctly.
# It allows for visual verification that the numeric features have been appropriately scaled.
print("First few rows of the dataset after normalization and standardization:\n", df.head())  # Print the first few rows to verify changes.

# Save the normalized and standardized dataset
# Justification: Saving the processed dataset ensures that the normalized and standardized data is preserved for subsequent steps in the pipeline.
# This step is crucial for reproducibility and for ensuring that the correct dataset is used in further analysis or modeling.
output_path = '../Outputs_FeatureSelection/FeatureSelection_Prep2.csv'  # Define the path to save the processed dataset.
df.to_csv(output_path, index=False)  # Save the dataset without the index to maintain consistency in further analysis.

print(f"Dataset after normalization and standardization saved to {output_path}")  # Inform the user that the dataset has been saved.
```

```python
# Step 5: Feature Engineering and Handling Outliers 

import pandas as pd
import numpy as np

# Load the dataset from the previous step (Prep2)
# Justification: We start by loading the dataset that has been standardized and normalized. This ensures that all features 
# are on a consistent scale before proceeding with feature engineering and outlier handling.
input_path = '../Outputs_FeatureSelection/FeatureSelection_Prep2.csv'
df = pd.read_csv(input_path)

# Feature Engineering
# Justification: Feature engineering is critical to enhance the model's predictive power by creating new features that better
# capture the underlying patterns in the data. These new features can reveal interactions between existing features or emphasize 
# important characteristics that weren't initially apparent.

# Interaction terms (example)
# Justification: Creating interaction terms (products of features) can help capture the combined effect of two features,
# which may have a significant impact on the target variable when considered together, but not when considered individually.
df['Interaction_0_1'] = df['0'] * df['1']

# Polynomial features (example)
# Justification: Polynomial features can capture non-linear relationships between the features and the target variable.
# For instance, squaring a feature can highlight non-linear effects that a linear model might miss.
df['0_squared'] = df['0'] ** 2

# Extract datetime features if 'Time' column exists
# Justification: Extracting components from datetime fields like year, month, day, hour, and minute can reveal trends or 
# patterns related to time. For example, manufacturing processes might have different yields at different times of the day 
# or during different months.
if 'Time' in df.columns:
    df['Year'] = pd.to_datetime(df['Time']).dt.year
    df['Month'] = pd.to_datetime(df['Time']).dt.month
    df['Day'] = pd.to_datetime(df['Time']).dt.day
    df['Hour'] = pd.to_datetime(df['Time']).dt.hour
    df['Minute'] = pd.to_datetime(df['Time']).dt.minute
    df.drop(columns=['Time'], inplace=True)  # Drop the original 'Time' column after extracting useful features.

# Handling Outliers using the IQR method
# Justification: Outliers can significantly skew the results of statistical analyses and machine learning models. The Interquartile 
# Range (IQR) method is a robust technique to identify outliers based on the spread of the data, which makes it suitable for detecting
# extreme values that could negatively impact model performance.
# Exclude Pass/Fail from the outlier handling process
# Justification: 'Pass/Fail' is the target variable and should not be treated as a numeric feature for outlier detection, as it is categorical.
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = numeric_columns.drop('Pass/Fail')  # Exclude Pass/Fail from outlier handling.

def identify_outliers_iqr(data):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1  # Calculate the Interquartile Range (IQR)
    # Identify outliers as values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
    outliers = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
    return outliers

# Identify and cap outliers for numeric columns excluding Pass/Fail
# Justification: Capping outliers (replacing extreme values with the nearest value within the normal range) reduces the impact of 
# outliers on the model while preserving the overall distribution of the data.
for column in numeric_columns:
    outliers = identify_outliers_iqr(df[column])  # Identify outliers using the IQR method.
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the outliers by setting them to the lower or upper bound, respectively.
    df[column] = np.where(outliers, np.where(df[column] < lower_bound, lower_bound, upper_bound), df[column])

# Display the first few rows of the processed data
# Justification: Viewing the first few rows after feature engineering and outlier handling provides a quick check to ensure 
# the transformations have been applied correctly. It allows for visual confirmation that new features were created and that outliers were handled as expected.
print("First few rows of the dataset after feature engineering and handling outliers:\n", df.head())

# Save the processed dataset after feature engineering and handling outliers
# Justification: Saving the dataset after this step ensures that all engineered features and outlier treatments are preserved
# for future modeling. This is important for maintaining consistency in the data pipeline.
output_path = '../Outputs_FeatureSelection/FeatureSelection_Prep3.csv'  # Define the path to save the processed dataset.
df.to_csv(output_path, index=False)  # Save the dataset without the index.

print(f"Dataset after feature engineering and handling outliers saved to {output_path}")  # Inform the user that the dataset has been saved.
```

```python
# Step 6: Correlation Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the previous step (Prep3_Corrected)
# Justification: We load the dataset that has undergone feature engineering and outlier handling. This ensures that the features 
# used in the correlation analysis are cleaned, standardized, and potentially more relevant after the feature engineering process.
input_path = '../Outputs_FeatureSelection/FeatureSelection_Prep3.csv'
df = pd.read_csv(input_path)

# Ensure that 'Pass/Fail' is treated as a numeric value for correlation
# Justification: The 'Pass/Fail' column, which is the target variable, must be converted to numeric form to compute correlations
# with other numeric features. This conversion allows the correlation function to correctly interpret 'Pass/Fail' as a numeric variable.
df['Pass/Fail'] = pd.to_numeric(df['Pass/Fail'], errors='coerce')

# Drop rows where 'Pass/Fail' is NaN after conversion
# Justification: If any rows have NaN values in 'Pass/Fail' after conversion, they should be dropped to avoid errors or 
# misleading results in the correlation analysis. This ensures that the analysis is based on valid data.
df = df.dropna(subset=['Pass/Fail'])

# Select only numeric columns for correlation
# Justification: Correlation analysis requires numeric data. By selecting only the numeric columns, we exclude any categorical 
# or non-numeric data that could distort the correlation results.
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation with Pass/Fail
# Justification: Calculating the Pearson correlation coefficient for each feature with 'Pass/Fail' helps identify which features
# are most strongly associated with the target variable. This is crucial for feature selection in the modeling stage.
correlation_with_target = numeric_df.corr()['Pass/Fail'].sort_values(ascending=False)
print("Correlation of Features with Pass/Fail:\n", correlation_with_target)

# Visualize the top correlated features
# Justification: Visualizing the top correlated features provides a clear understanding of which features are most closely 
# associated with 'Pass/Fail'. A heatmap of the correlations among these top features can also reveal potential multicollinearity,
# which is important to address before modeling.
top_features = correlation_with_target.head(20).index  # Select the top 20 features most correlated with 'Pass/Fail'
plt.figure(figsize=(10, 8))  # Set the figure size for better readability
sns.heatmap(numeric_df[top_features].corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Top Features Correlated with Pass/Fail')
plt.show()
```

DATA MODELING STAGE

```python
# Step 7: Feature Selection and Initial Modeling with Class Imbalance Handling

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import joblib

# Load the dataset from the previous step (Prep3_Corrected)
# Justification: We load the processed dataset after handling missing values, normalization, feature engineering, 
# and outlier handling. This ensures that the features selected for modeling are clean and relevant.
input_path = '../Outputs_FeatureSelection/FeatureSelection_Prep3.csv'
df = pd.read_csv(input_path)

# Feature Selection: Select the top correlated features
# Justification: Based on the previous correlation analysis, we select features that have a strong correlation with the target variable (Pass/Fail).
# These features are likely to be predictive and useful in the modeling process, reducing the dimensionality and focusing the model on the most relevant data.
selected_features = [
    '59', '103', '247', '519', '510', '477', '205', '341', '348', '129', 
    '287', '152', '33', '385', '124', '425', '210', '298', '290'
]

# Add 'Pass/Fail' as the target variable
# Justification: The target variable 'Pass/Fail' is crucial for training the model as it represents the outcome we are trying to predict.
selected_features.append('Pass/Fail')

# Subset the dataframe to only include selected features
# Justification: By subsetting the dataframe to include only the selected features and the target variable, 
# we focus the model on the most relevant data, improving efficiency and potentially the accuracy of the model.
df_selected = df[selected_features]

# Split the data into training and testing sets
# Justification: Splitting the data allows us to train the model on one part (training set) and evaluate its performance on another (testing set).
# The `stratify=y` parameter ensures that the split maintains the original distribution of the target variable, which is important in cases of class imbalance.
X = df_selected.drop('Pass/Fail', axis=1)  # Features
y = df_selected['Pass/Fail']  # Target variable

# Use a 70-30 split for training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Build a Logistic Regression model with class_weight set to 'balanced'
# Justification: Logistic Regression is chosen for its simplicity and interpretability, making it a good starting point for modeling.
# The `class_weight='balanced'` parameter addresses the class imbalance by adjusting the weights inversely proportional to class frequencies,
# ensuring that the model does not become biased towards the majority class.
model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

# Make predictions
# Justification: After training, the model is used to predict the outcomes on the testing set, which allows us to evaluate its performance.
y_pred = model.predict(X_test)

# Evaluate model performance
# Justification: Multiple metrics are used to evaluate the model:
# - Accuracy: The proportion of correct predictions out of all predictions.
# - Precision: The proportion of true positive predictions among all positive predictions.
# - Recall: The proportion of true positive predictions among all actual positives.
# - F1 Score: The harmonic mean of precision and recall, providing a single metric that balances both concerns.
# - ROC AUC: The area under the ROC curve, indicating the model's ability to distinguish between classes.
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=1, zero_division=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

# Display evaluation metrics
# Justification: Printing the metrics provides a clear understanding of the model's performance, helping us to decide if further tuning or a different model is needed.
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")

# Save the selected features dataset for future use
# Justification: Saving the dataset ensures that the exact subset of data used for modeling is preserved,
# making it easier to reproduce results or conduct further analysis later.
output_path = '../Outputs_FeatureSelection/FeatureSelection_Modeling1.csv'
df_selected.to_csv(output_path, index=False)

print(f"Dataset with selected features saved to {output_path}")

# Save the trained model for future use
# Justification: Saving the model allows for easy reuse without needing to retrain. This is useful for deployment or further evaluation.
model_filename = '../Outputs_FeatureSelection/best_feature_selection_model.pkl'
joblib.dump(model, model_filename)

print(f"Model saved to {model_filename}")
```

EVALUATION STAGE

```python
# Step 8: Model Evaluation

import pandas as pd
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import joblib

# Load the trained model and the selected features dataset
# Justification: We load the model that was trained and saved in the previous step. 
# This allows us to evaluate the model's performance on the dataset with the selected features, ensuring consistency.
model_path = '../Outputs_FeatureSelection/best_feature_selection_model.pkl'
model = joblib.load(model_path)

data_path = '../Outputs_FeatureSelection/FeatureSelection_Modeling1.csv'
df_selected = pd.read_csv(data_path)

# Split the data into features and target
# Justification: We separate the dataset into features (X) and the target variable (y) to facilitate predictions and evaluation.
X = df_selected.drop('Pass/Fail', axis=1)
y = df_selected['Pass/Fail']

# Make predictions
# Justification: Using the trained model, we generate predictions (y_pred) and prediction probabilities (y_prob) for the evaluation dataset.
y_pred = model.predict(X)
y_prob = model.predict_proba(X)[:, 1]

# Confusion Matrix
# Justification: The confusion matrix provides a detailed breakdown of the model's performance, showing the number of true positives, true negatives, 
# false positives, and false negatives. This is useful for understanding the model's strengths and weaknesses.
conf_matrix = confusion_matrix(y, y_pred)
plt.figure(figsize=(5, 4))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
plt.xticks([0, 1], ['-1', '1'])  # Labels for the classes
plt.yticks([0, 1], ['-1', '1'])  # Labels for the classes
plt.xlabel('Predicted label')
plt.ylabel('True label')
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(j, i, format(conf_matrix[i, j], 'd'), ha="center", va="center")
plt.show()

# ROC Curve
# Justification: The ROC curve is a graphical representation of the model's diagnostic ability. 
# It plots the true positive rate (sensitivity) against the false positive rate (1-specificity) for different threshold values. 
# The area under the ROC curve (AUC) is a single scalar value that provides a summary of the model's performance.
fpr, tpr, _ = roc_curve(y, y_prob, pos_label=1)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(5, 4))
plt.plot(fpr, tpr, color='orange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line representing random guess
plt.xlim([0.0, 1.0])  # Set the limits for the x-axis
plt.ylim([0.0, 1.05])  # Set the limits for the y-axis
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Precision-Recall Curve
# Justification: The precision-recall curve is particularly useful in the context of imbalanced datasets. 
# It shows the trade-off between precision and recall for different thresholds, helping to evaluate how well the model performs across different levels of recall.
precision, recall, _ = precision_recall_curve(y, y_prob, pos_label=1)
plt.figure(figsize=(5, 4))
plt.plot(recall, precision, color='purple', lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
